```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()

import matplotlib.pyplot as plt
%matplotlib inline

# To Check iris dataset
dir(iris)
```

```python
# To Check the features name
iris.feature_names
```

```python
# To Check the target variable(dependent variable)
iris.target
```

```python
# To Check the target variable(dependent variable)
# 0(setosa) 1(versicolo) 2 (virginica')
iris.target_names
```

```python
# To Check the data which is 2D array
iris.data
```

```python
# To Convert into dataframe
df = pd.DataFrame(iris.data,columns = iris.feature_names)
df.head()
```

```python
# To add Target column
df['target'] = iris.target
df.head()
```

```python
df[df['target'] == 1].head()
```

```python
df[df['target'].isin([0,1,2])]
```

```python
# To add new column
df['flower_name'] = df.target.apply(lambda x : iris.target_names[x])
df.head()
```

```python
df[35:60]
```

```python
# To create 3 diffrent datasets
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color = 'green',marker = '+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color = 'blue',marker = '.')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color = 'red',marker = '+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color = 'black',marker = '.')
plt.xlabel('petal length')
plt.ylabel('petal width')
```

```python
# import train test split
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target','flower_name'],axis = 'columns') # It will drop both dependet variables.
x.head()
```

```python
y = df.target
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
len(x_train)
# len(x_test)
# len(y_train)
# len(y_test)
```

```python
# To create SVM Model
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
# To check the model score
model.score(x_test,y_test)
```

```python
# sample of x_test to predict some value
x_test
```

```python
model.predict([[7.7,	2.8,	6.7,	2.0]])
```

```python
model_R = SVC(C = 20)
model_R.fit(x_train,y_train)
model_R.score(x_test,y_test)
```

```python
model_G = SVC(gamma = 10)
model_G.fit(x_train,y_train)
model_G.score(x_test,y_test)
```

```python
model.linear_kernel=SVC(kernel = 'linear')
model.linear_kernel.fit(x_train,y_train)
```

```python
model.linear_kernel.score(x_test,y_test)
```
